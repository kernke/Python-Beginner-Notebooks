```python
def line(p0,p1,N):
    xcoord=np.zeros(N)
    ycoord=np.zeros(N)
    xstep=(p1[0]-p0[0])/N
    ystep=(p1[1]-p0[1])/N
    step_width=np.sqrt(xstep**2+ystep**2)
    for i in range(N):
        xcoord[i]=p0[0]+i*xstep
        ycoord[i]=p0[1]+i*ystep
    return xcoord,ycoord,step_width
```

```python
xcoord,ycoord,step=line(p0,p1,100)
```

```python
def line_profile_nearest(X,Y,Z,p0,p1,N):
    xcoord,ycoord,step=line(p0,p1,N)
    line_points=list(zip(xcoord,ycoord))
    
    Xflat=np.ravel(X)
    Yflat=np.ravel(Y)
    serialized_coords=list(zip(Xflat,Yflat))
    serialized_image=np.ravel(Z)
    
    distance_matrix=cdist(line_points,serialized_coords)
    closest_point_indices=np.argmin(distance_matrix,axis=-1)
    
    profile=serialized_image[closest_point_indices]
    profile_x=np.arange(len(profile))*step
    
    return profile_x,profile
```

```python
def bandpass(image,low,high,strength):
    
    f = np.fft.fft2(Z)
    fshift = np.fft.fftshift(f)

    mask=np.ones(image.shape)
    center=(np.array(np.shape(image))-1)/2
    circ_coord_low=skimage.draw.disk(center,low)
    circ_coord_high=skimage.draw.disk(center,high)
    for i in np.transpose(circ_coord_high):
        mask[i[0],i[1]] /= strength
    for i in np.transpose(circ_coord_low):
        mask[i[0],i[1]] *= strength
        
    f_filtered=fshift*mask
    
    f_back_shifted=np.fft.fftshift(f_filtered)
    reverted_image=np.real(np.fft.ifft2(f_back_shifted))
    return reverted_image,mask
```

# Code that produces the Excercise-Image and Solution-Spoiler

```python
X,Y=np.meshgrid(np.linspace(0,5,1000),np.linspace(0,5,2000))

def gauss(x,y,x0,y0,sigma,A):
    return A*np.exp(-((x-x0)**2+(y-y0)**2)/(2*sigma**2))
Z=gauss(X,Y,2,1.5,0.5,15) + gauss(X,Y,3.5,4,0.25,12)
Z+=np.random.normal(0,4,Z.shape)

Z=Z-np.min(Z)
Z=Z/np.max(Z)*255
cv2.imwrite("example_img.png",Z.astype(np.uint8))
```

```python
closest_point_indices=np.argmin(distance_matrix,axis=-1)
for i in range(len(closest_point_indices)):
    distance_matrix[i,closest_point_indices[i]]=np.inf
second_closest_point_indices=np.argmin(distance_matrix,axis=-1)

def line_profile_linear(X,Y,Z,p0,p1,N):
    xcoord,ycoord,step=line(p0,p1,N)
    line_points=list(zip(xcoord,ycoord))
    Xflat=np.ravel(X)
    Yflat=np.ravel(Y)
    serialized_coords=list(zip(Xflat,Yflat))
    serialized_image=np.ravel(Z)
    distance_matrix=cdist(line_points,serialized_coords)
    
    profile=np.zeros(N)
    weight=np.zeros(N)
    for i in range(3):
        closest_point_indices=np.argmin(distance_matrix,axis=-1)
        values=serialized_image[closest_point_indices]
        for j in range(len(closest_point_indices)):
            profile[j] +=values[j]/distance_matrix[j,closest_point_indices[j]]
            weight[j] += 1/distance_matrix[j,closest_point_indices[j]]
            distance_matrix[j,closest_point_indices[j]]=np.inf
    profile/=weight
    
    profile_x=np.arange(len(profile))*step
    return profile_x,profile

profile_x,profile=line_profile_linear(newX,newY,example,p0,p1,100)
plt.plot(profile_x,profile)
```
